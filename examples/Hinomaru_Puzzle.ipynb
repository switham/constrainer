### Breaking Down the Problem

The `constrainer` module was written to solve puzzles.  Using it amounts to analyzing the problem, writing code to feed it into an API, calling the solution generator, and writing code to present the results.  Perhaps it's anticlimactic to have the glory occur inside a black box (of course I invite you to read and modify `constrainer`'s source), but I like finding out whether new puzzles can be fitted to this minimalist framework.  This time there was a new wrinkle that will be explained later.  But first the raw data needs to be distilled.

In John's hands, the puzzle has already been reduced quite a bit, to a 12 x 18 grid of ones and zeroes, and we need to fill it with twelve tiles, each tile being either 3 x 6 or 6 x 3 depending on how it's turned.

All of the sizes are divisible by three, so the tiles can only be placed on coordinates that are divisible by three, making it more like a 4 x 6 grid of 3 x 3... call them **squares**, with each tile being 1 x 2 (or 2 x 1) squares.  Each tile has two **faces**, and each face can only go in a place where it matches the 3 x 6 or 6 x 3 pattern of ones and zeroes.

Square locations will be represented by (row, column) two-tuples, with `0 <= row < 4`, `0 <= column < 6`, and `(0, 0)` the top-left corner.  A **place** is a square location followed by either "across" or "down", e.g. `((1, 2), "down")`.

The code here includes a bit of testing or sanity-checking.

```python
SQUARES = set((row, col) for row in range(4) for col in range(6))

PLACES = set(((row, col), "down") for row in range(3) for col in range(6))  \
       | set(((row, col), "across") for row in range(4) for col in range(5))
# "|" is one way to get the set union.

print len(SQUARES), "squares,", len(PLACES), "places."

def place_squares(place):
    assert place in PLACES
    
    (row, col), orientation = place
    if orientation == "across":
        return (row, col), (row, col + 1)
    else:
        return (row, col), (row + 1, col)

square_counts = dict((square, 0) for square in SQUARES)
for place in PLACES:
    if not all(square in SQUARES for square in place_squares(place)):
        print place, "goes off the edge:", place_squares(place), "!"
        break
        
    for square in place_squares(place):
        square_counts[square] += 1

    square, orientation = place
    if orientation not in ("down", "across"):
        print place, "isn't a proper place!"
        break
else:
    print "All the places are well-formed."
    for n in sorted(set(square_counts.values())):
        matching = [s for s in SQUARES if square_counts[s] == n]
        print len(matching), "squares are part of", n, "places."
```

Python can compare tuples for equality (also for "<" and ">") and use them as keys for dictionaries, so instead of giving a face a name or number we will just represent it by a 3 x 6 tuple-of-tuples.  But faces need to be recognized as the same if they are rotated 180&deg;, so we will always treat the rotation that Python considers "less" as the canonical one.  In order to rotate faces, we will convert them to numpy arrays, then back to tuples-of-tuples.  Numpy has a nice function **rot90()** to rotate a 2D array.  Numpy can also cut a rectangular **slice** out of a 2D array (e.g., from FLAG_DATA).

```python
def face_to_np(face):
    assert type(face) == tuple
    assert np_shape(face) == (3, 6)
    
    return np.array(face, dtype=np.int8)

def np_to_face(np_face):
    assert type(np_face) == np.ndarray
    assert np_face.shape == (3, 6)
    
    return tuple(tuple(row) for row in np_face)

def rot180(np_face):
    return np.rot90(np.rot90(np_face))

def np_to_canonical_face(np_face):
    f1 = np_to_face(np_face)
    f2 = np_to_face(rot180(np_face))
    return min(f1, f2)

def canonical_face(face):
    return np_to_canonical_face(face_to_np(face))

def place_slice(place):
    """
    Return a 3 x 6 slice out of FLAG_ARRAY.
    Return the slice oriented in the "across" way
    even if the place is "down", 
    i.e. rotate it 90 degrees if necessary.
    """
    assert place in PLACES
    
    (row, col), orientation = place
    f_row, f_col = row * 3, col * 3
    if orientation == "across":
        return FLAG_ARRAY[f_row : f_row + 3, f_col : f_col + 6]
    else:
        return np.rot90(FLAG_ARRAY[f_row : f_row + 6, f_col : f_col + 3])
    
def place_face(place):
    """
    Return the canonical face that matches place on the flag.
    """
    return np_to_canonical_face(place_slice(place))

# Convert the TILES_DATA to use canonical face orientations.

for i, (side1, side2) in enumerate(TILES_DATA):
    TILES_DATA[i] = tuple(sorted( (canonical_face(side1), canonical_face(side2)) ))
    
FACES = set(face for tile in TILES_DATA for face in tile)

def face_fits_place(face, place):
    assert face in FACES, "face must be canonical for match to work."
    return face == place_face(place)
    
print "TILES_DATA now has", len(FACES), "distinct faces."

for tile in TILES_DATA:
    side1, side2 = tile
    if side1 == side2:
        print "This tile is the same on both sides!"
        print tile
        break        
else:
    print "No tiles are the same on both sides."
    
if len(TILES_DATA) == len(set(TILES_DATA)):
    print "There are no duplicate tiles."
else:
    print "There are duplicate tiles!"

for face in FACES:
    if not any(face_fits_place(face, place) for place in PLACES):
        print "This face doesn't fit anyplace!"
        print face_to_np(face)
        print
        break
else:
    print "Every face has a place."
        
for place in PLACES:
    if not place_face(place) in FACES:
        print "This place doesn't have a matching face!"
        print place_slice(place)
        print
        break
else:
    print "Every place has a matching face."

count_face_place = sum(face_fits_place(face, place) 
                       for face in FACES for place in PLACES)
print count_face_place, "matches between", len(FACES), "faces and", len(PLACES), "places"
```

Here are the facts about the puzzle:

* A given place covers a certain two squares.
* A square can have two places that overlap it (at a corner), three (on an edge), or four (in the middle area), but in a solution, there can only be a tile in one of those places. 
* Only one face fits in a given place.
* Many (>= 2) places fit a given face.
* A tile shows exactly one of its two faces.
* A given face shows on exactly as many tiles as places it is placed on the board (hmm...)

The program will convert the puzzle data onto punch cards that the [constrainer](https://github.com/switham/constrainer) library accepts, press the Go button, and then `constrainer` will spit out solutions.  Actually `constrainer` uses Python objects in three classes for its inputs:

A **State** ties the whole problem together and has the methods that do the work.

A **BoolVar** is like a checkbox that is originally empty (actually set to *Maybe*), but eventually gets set to *True* or *False*.  All of the choices in the problem have to be represented with BoolVars.

A **BoolConstraint** is a rule that says, out of some set of BoolVars, at least *this* many, and at most *this* many, have to be true.  For instance, this...

    state = State()
    a = BoolVar(state)
    b = BoolVar(state)
    BoolConstraint(state, a, b, min_True=1, max_True=1)
    
...means that exactly one out of {*a*, *b*} must be *True*.  In other words, once we find out or guess one, the other must have the opposite truth value.

There's more information in the [constrainer Readme](https://github.com/switham/constrainer/blob/master/README) file, but the general idea is that BoolVars represent choices that the program needs to make, and BoolConstraints represent the consistency rules between choices.  Usually, vars are about the relations between things (for instance, *This piece is in that place,* while constraints are about a given thing (for instance, *There can only be one thing in this place*).

Both the BoolVars and BoolConstraints will be stored in dictionaries, some in more than one dictionary.  One set of vars are in both **`tile_shows_face[tile][face]`** and **`face_shown_vars[face]`**, the other set in **`place_is_used[place]`** (i.e. has a tile on it, true or false).  That's enough vars to specify a whole solution, because if a place is used, we know what face it will be, and what tile(s) show(s) that face.

The constraints are **`shows_one_face[tile]`**, **`under_one_tile[square]`**, and **`shows_eq_places[face]`** (explained later).

```python
def init_tile_face():
    # Each tile shows one of two faces.
    tile_shows_face = dict()
    face_shown_vars = dict( (face, []) for face in FACES)
    shows_one_face = dict()
    for i, tile in enumerate(TILES_DATA):
        tile_vars = [BoolVar(state, i=i, tile=tile, face=face)
                     for face in tile]
        tile_shows_face[tile] = dict( (v.face, v) for v in tile_vars)
        for v in tile_vars:
            face_shown_vars[v.face].append(v)
        shows_one_face[tile] = BoolConstraint(state, *tile_vars, min_True=1, max_True=1)    
    return tile_shows_face, shows_one_face, face_shown_vars

def init_square_place():
    # Each square is covered by (a tile in) one of two places.
    # Each place is either used or not.
    sq_constraints = [BoolConstraint(state, square=square, min_True=1, max_True=1)
                      for square in SQUARES]
    under_one_tile = dict( (c.square, c) for c in sq_constraints)
    place_is_used = dict()
    for place in PLACES:
        place_is_used[place] = BoolVar(state, place=place)
        for square in place_squares(place):
            under_one_tile[square].constrain(place_is_used[place])
    return under_one_tile, place_is_used
```

The 

`shows_eq_places[face]` constraint means this.  If there are (say) three tiles that can show a given face, then looking at the tiles you might think that between zero and three instances of that face will be showing.  Similarly, if there are four places on the puzzle where a tile with that face might fit, then looking at the board you might think that between zero and four of those places will have a tile in them (as opposed to being covered by halves of two different tiles).  But clearly, in the end, as many tiles will be showing that face as tiles showing that face will be in eligible places for that place on the board.

(Immediately that constraint would rule out four places showing that face, without having to treat the 3 vs. 4 mismatch as a special case.)

What this does is prevent the program from spinning through all the six different ways of arranging three same-looking tiles in three similar places, and the same for any other face with multiple tiles.  It cuts down on the combinatorial explosion.  We don't really care *which* same-looking tiles are in which of those places, although we do care which group of tiles will be used for it.  The search isn't multiplied by possibilites we don't represent.

To output a "solution", though, the program will either have to pick arbitrary tile-place assignments, or describe sets of tiles going to sets of places.

Constraining two sets of BoolVars to have the same number of *True*s is a new feature for `constrainer`.  It's done in the new function `constrain_same_n_true()`.

```python
from sys import stdout
from time import clock

def constrain_same_n_true(state, vars1, vars2, **kwargs):
    # Create one new var to be the opposite of each var in vars1.
    inv_vars1 = [BoolVar(state, var=var, name="NOT(%s)" % var)
                for var in vars1]
    
    # Constrain each var to be the opposite of its counterpart:
    for ivar in inv_vars1:
        BoolConstraint(state, ivar, ivar.var, min_True=1, max_True=1)
    # Now constrain the union of inv_vars1 and vars2 to have
    # as many Trues as there are members of inv_vars1.  Then,
    # All of vars1 and vars2 False
    #     means all of inv_vars1 True,
    #     and the constraint is satisfied.
    # and whenever a vars1 var turns True,
    #     the corresponding inv_vars1 var turns False,
    #     and a vars2 var must turn True to keep the balance.
    n = len(inv_vars1)
    return BoolConstraint(state, *(inv_vars1 + vars2), 
                          min_True=n, max_True=n, **kwargs)

def init_shows_eq_places():
    # Each face is shown by the same number of tiles
    # as there are places in use that the face fits.
    shows_eq_places = dict()
    for face in FACES:
        shown_vars = face_shown_vars[face]
        placed_vars = [v for (place, v) in place_is_used.iteritems()
                       if face_fits_place(face, place)]
        shows_eq_places[face] = constrain_same_n_true(state, shown_vars, placed_vars)
    return shows_eq_places

def init_constrainer():
    global state
    global tile_shows_face, place_is_used, face_shown_vars
    global shows_one_face, under_one_tile, shows_eq_places
    
    state = State()
    tile_shows_face, shows_one_face, face_shown_vars = init_tile_face()
    under_one_tile, place_is_used = init_square_place()
    shows_eq_places = init_shows_eq_places()
    
init_constrainer()
print "Initialized."
```

#### Printing Solutions

Here's the code for printing the solution once `constrainer` finds it, based on values of BoolVars.
The `solution_pic()` function produces a typewriter picture of the board with tiles on it,
and the tiles showing which side of each tile goes to which location on the board.  On the placement
instructions, rows and columns are numbered starting from 1.

Typewriter pictures are represented as lists of strings (one per line), and a lot of this section
is just about producing and pasting together "pix".

```python
def solution_square_places():
    """
    returns a dict like {square: place, ...}.
    """
    result = dict()
    for place in PLACES:
        if place_is_used[place]:
            for square in place_squares(place):
                result[square] = place
    return result

def solution_tile_placements():
    """
    returns a dict like {tile: (face_index, place), ...},
    where face_index is 0 or 1 for which side of the tile.
    Identical-faced tiles are distributed among matching places arbitrarily.
    """
    face_places = defaultdict(list)
    for place in PLACES:
        if place_is_used[place]:
            face_places[place_face(place)].append(place)
            
    tile_placements = dict()
    for tile in TILES_DATA:
        face_index = int(tile_shows_face[tile][tile[1]].value)
        face = tile[face_index]
        tile_placements[tile] = (face_index, face_places[face].pop(0))
    return tile_placements

def square_slice(square):
    """ A 3 x 3 array of ones and zeroes from a square of the flag. """
    assert square in SQUARES
    
    (row, col) = square
    f_row, f_col = row * 3, col * 3
    return FLAG_ARRAY[f_row : f_row + 3, f_col : f_col + 3]

def square_pic(square, sqp, left_edge=False, top_edge=False):
    """
    Given square (row, col), return a typewriter picture of what's there.
    sqp is the output of solution_square_places.
    Output example:
       ["      |",
        "    ##|",
        "######|",
        "------+",]
    The lines on the right or bottom become extensions
    (duplicates) of the neighboring " " or "#" if a tile
    extends across the corresponding division.
    left_edge says whether to *insert* a vertical border character per line.
    top_edge says whether to add an initial horizontal border line.
    """
    assert square in SQUARES
    
    row, col = square
    right_edge = col >= 5 or sqp[square] != sqp[(row, col + 1)]
    bottom_edge = row >= 3 or sqp[square] != sqp[(row + 1, col)]
    ss = square_slice(square)
    assert ss.shape == (3, 3)
    return array_pic(ss, right_edge, bottom_edge, left_edge, top_edge)

def array_pic(ar, right_edge, bottom_edge, left_edge=False, top_edge=False):
    """
    Return a rectangular typewriter picture based on a 2D bool array.
    ar is a 2D numpy array, tuple of tuples or list of lists.
    Output example:
       ["      |",
        "    ##|",
        "######|",
        "------+",]
    left_edge and top_edge *insert* edges (not present in the example).
    right_edge and bottom_edge say whether to draw edge lines.
    When a right or bottom edge is not present it's replaced by 
    duplicates of the characters to the left or above.
    """
    lc = "+" if left_edge else ""
    left = "|" if left_edge else ""
    if top_edge:
        pic = [lc + "--" * len(ar[0]) + "+"]
    else:
        pic = []        
    for row, row_data in enumerate(ar):
        line = "".join("##" if bit else "  " for bit in row_data)
        line += "|" if right_edge else line[-1]
        pic.append(left + line)
    pic.append(lc + ("--" * len(row_data) if bottom_edge else line[:-1]) + "+")
    return pic

def pad_pic_height(pic, height):
    """ Add blank lines to pic to pad it to height. """
    pad = height - len(pic)
    return pic if pad <= 0 else pic + [" "] * pad

def pad_pic_heights(pix, min_height=0):
    """
    Pad pics in pix to have identical heights.
    The vertical padding is not horizontally padded!
    Do pad_pic_widths after pad_pic_heights, or use pad_pic_heights_widths.
    """
    height = max(min_height, max(len(pic) for pic in pix))
    if min(len(pic) for pic in pix) == height:
        return pix
    else:
        return [pad_pic_height(pic, height) for pic in pix]

def pad_pic_width(pic, min_width=0):
    """ Pad pic to have uniform width within itself. """
    width = max(max(len(line) for line in pic), min_width)
    return [line.ljust(width) for line in pic]

def pad_pic_widths(pix, min_width=0):
    """
    Pad each pic to be rectangular, not all to the same width.
    No vertical padding is done.
    Do pad_pic_heights before pad_pic_widths, or use pad_pic_heights_widths.
    """
    return [pad_pic_width(pic, min_width) for pic in pix]

def pad_pic_heights_widths(pix, min_height=0, min_width=0):
    """
    Make each pic rectangular and all pics have the same height.
    """
    return pad_pic_widths(pad_pic_heights(pix, min_height), min_width)

def join_pix_LR(pix, min_height=0, min_width=0):
    pix = pad_pic_heights_widths(pix, min_height, min_width)
    height = len(pix[0])
    return ["".join(pic[i] for pic in pix) for i in range(height)]

def board_pic(sqp):
    """
    Return a typewriter picture of the current puzzle solution.
    sqp is the output of solution_square_places.
    """
    result = []
    for row in range(4):
        row_pix = [square_pic((row, col), sqp, top_edge=(row == 0), left_edge=(col == 0))
                   for col in range(6)]
        result += join_pix_LR(row_pix)
    return result

def tile_pic(tile, top_edge):
    assert shape(tile) == (2, 3, 6)
    
    side_pix = []
    for i, face in enumerate(tile):
        left_edge = (i == 0)
        side_pix.append(array_pic(face, True, True, left_edge, top_edge))
    return join_pix_LR(side_pix)

def placement_pic(place, top_edge):
    (row, col), orientation = place
    pic = [" "] if top_edge else []
    pic += [
        "row %d" % (row + 1),
        "col %d" % (col + 1),
        orientation,
        ]
    return pad_pic_width(pic, 6)

def tile_placement_pic(tiles, stp):
    """
    Pic of a block of tiles with placement instructions.
    The placement instructions go next to the side of the tile that shows.
    The block has a top & bottom line, & lines but not gaps between tiles.
    stp is the output of solution_tile_placements,
    a dict like {tile: (face_index, place), ...}.
    """
    pic = []
    for i, tile in enumerate(tiles):
        which_side, place = stp[tile]
        top_edge = (i == 0)
        tpic = tile_pic(tile, top_edge)
        ppic = placement_pic(place, top_edge)
        slug = [" ".ljust(len(ppic[0]))]
        sp = [" "]
        if which_side == 0:
            pic += join_pix_LR([sp, ppic, sp, tpic, sp, slug, sp])
        else:
            pic += join_pix_LR([sp, slug, sp, tpic, sp, ppic, sp])
    return pic

def join_list(spacer, things):
    """
    New list with spacer inserted between adjacent things.
    E.g.
        join_list(0, [1, 2, 3]) => [1, 0, 2, 0, 3]
    """
    result = things[:1]
    for thing in things[1:]:
        result.append(spacer)
        result.append(thing)
    return result

def solution_pic():
    ssp = solution_square_places()
    stp = solution_tile_placements()
    bp = board_pic(ssp)
    blocks = [bp]
    pic = []
    for i in range(0, len(TILES_DATA), 4):
        tiles = TILES_DATA[i : i + 4]
        blocks.append(tile_placement_pic(tiles, stp))
    for i in range(0, len(blocks), 2):
        pic += join_pix_LR(join_list([" "], blocks[i : i + 2]))
    return pic
```
